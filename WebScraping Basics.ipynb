```python
# create the soup of the contents obtained
soup = BeautifulSoup(data,'html.parser')
```

```python
# grab hold of all the children sites
children_sites = []
for li in soup.findChildren('li'):
    for a in li.find_all('a'):
        print (a['href'])
        children_sites.append((a['href']))
```

```python
# we are interested in the skindeep products which have been put into category
# grab the path (relative to the base url) for the skindeep products
new_links = []
for sites in children_sites:
    if '/skindeep/browse/' in sites:
        print (sites)
        new_links.append(base_url.format(sites))
```

```python
# get the content (product links, product name etc from the new links obtained above)
# print out the link number only if the hit was success
url_connection_response = {}
link_number = 0
for urls in new_links:
    print (f"Link number: {link_number}")
    response = requests.get(urls)
    # status code = 200 means a success
    # storing the entire content into a dictionary
    if response.status_code == 200:
        url_connection_response[urls] = response
    link_number += 1
```

```python
import pandas as pd
class scrape_ewg_link(object):
    
    def __init__(self,link):
        self.link = link
        self.response = requests.get(self.link)
        self.status = self.response.status_code
        if self.status == 200:
            self.soup = BeautifulSoup(self.response.content,'html.parser')
        else:
            print (f"Check Link: {link}")
        self.products_dataframe = pd.DataFrame()
        
    def get_product_links(self):
        links_list = []
        for product_tile in self.soup.find_all('div',attrs={"class":"product-tile"}):
            for a_href in product_tile.find_all('a'):
                links_list.append(a_href['href'])
        self.products_dataframe['Links'] = links_list

    def get_product_names(self):
        name_list = []
        for product_tile in self.soup.find_all('p',attrs={"class":"product-name"}):
            for a_href in product_tile.find_all('a'):
                name_list.append(a_href.text)
        self.products_dataframe['Product_Name'] = name_list
    
    def get_product_company_names(self):
        company_list = []
        for product_tile in self.soup.find_all('p',attrs={"class":"product-company"}):
            for a_href in product_tile.find_all('a'):
                company_list.append(a_href.text)
        self.products_dataframe['Company_Name'] = company_list
        
    
    def main(self):
        self.get_product_names()
        self.get_product_company_names()
        self.get_product_links()
        return self.products_dataframe
```

```python
# 
dataframe_collector = {}
for index, links in enumerate(new_links):
    print (f"Link Number: ",index)
    dataframe_collector[links] = scrape_ewg_link(links).main()
```

```python
dataframe_collector
```

```python
# convert the dictionary to a dataframe with product_names, company_names, links etc
final = pd.DataFrame()
for links in dataframe_collector:
   # print(type(dataframe_collector[links]))
    final = pd.concat([final,dataframe_collector[links]],ignore_index=True,sort=False)
final
```

```python
# create a new scraping routine to obtain information for each product using the links extracted above :D
class scrape_product_info(object):
    
    def __init__(self,link):
        self.link = link
        self.response = requests.get(link)
        print (self.response.status_code)
        self.soup = BeautifulSoup(self.response.content,'html.parser')
        self.heads = []
        self.table = pd.DataFrame()
        
    def get_table_heads(self):        
        for thead in self.soup.find_all("thead"):
            for th in thead.find_all("th"):
                if ' ' not in th: 
                    self.heads.append(th.text)
        self.table = pd.DataFrame(columns=self.heads)
        
    def get_table_values(self):
        bs = BeautifulSoup(self.response.text,'lxml')
        table_body=bs.find('tbody')
        rows = table_body.find_all('tr')
        for index,row in enumerate(rows):
            cols=row.find_all('td')
            cols=[x.text.strip() for x in cols]
            self.table.loc[index,:] = cols
    
    def get_animal_testing_policies(self):
        policy = self.soup.find("section",attrs={"id":"animal-testing-policies"})
        print (policy.find('h3').text)     
        
    def get_label_information(self):
        label_info = self.soup.find("section",attrs={"id":"label-information"})
        print ('Ingredients:', {label_info.find("p").text})
        label_info_h = label_info.find_all('h3')
        try:
            print ("Directions from packing: ",label_info_h[0].find('p'))    
        except:
            print ("Directions from packing: ",None)    
        try:
            print ("Warnings from packing: ",label_info_h[1].find('p'))
        except:
            print ("Warnings from packing: ",None)
    
    def get_product_image(self):
        self.image = self.soup.find('div',attrs={"class":"product-image"}).find('img').attrs['src']
        return self.image
    
    def get_ingredient_concern(self):
        for image in self.soup.find_all("div",attrs={"class":"thebigimage"}):
            print (image.find("div",attrs={"class":"gauge-img-wrapper"}).find("img").attrs['alt'])
            
    def get_other_ingredient_concern(self):
        for image in self.soup.find_all("div",attrs={"class":"thebigimage"}):
            print (image.find("div",attrs={"class":"gauge-img-wrapper"}).find("img").attrs['alt'])
    
    def get_other_high_concerns(self):
        concerns = self.soup.find('section',attrs={"id":"other-concerns"}).find_all('ul')
        if len(concerns) > 0:
            for concern in concerns[0].find_all('li'):
                print (concern.text)
            
    def get_other_low_concerns(self):
        concerns = self.soup.find('section',attrs={"id":"other-concerns"}).find_all('ul')
        if len(concerns)>1:
            for concern in concerns[1].find_all('li'):
                print (concern.text)
                
    def get_score_information(self):
        score_list = []
        for td_class in self.soup.find_all('img',attrs={'class':"ingredient-score score-popup"}):
            score_link = td_class.attrs['src'].split('-')
            for index, value in enumerate(score_link):
                if value == 'static.ewg.org/skindeep_rails/score':
                    score_list.append(score_link[index+1])
        self.table.Score = score_list
        
    def main(self):
        self.get_table_heads()
        self.get_table_values()
        self.get_score_information()
        self.table.head()
```
